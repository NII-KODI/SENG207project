```python
#Plotting a Histogram to analyse Alcohol Content Distribution
dataset.hist()
plt.figure(figsize=(8,6))
plt.show()
plt.savefig('red_wine_histogram.png')
plt.show()

         
```

```python
datasetplot=dataset.plot(kind='hist',bins=35)
datasetplot.set_title("Red Wine Content Distribution")
datasetplot.set_xlabel("Red wine content")
plt.show()
```

```python
#Creating a Bar Plot for dataset to check the pH frequency of fixed acidity
dataset.plot(kind='bar',x='citric acid',y='pH',legend=None)
plt.xlabel('Citric Acid')
plt.ylabel('pH')
plt.show()
```

```python
dataset.plot(kind='box',x='alcohol',y='quality',legend=None)
plt.title('Alcohol Content vs Quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show()
```

```python
plt.scatter(dataset['alcohol'],dataset['pH'])
plt.title('Alcohol Content vs pH Level')
plt.xlabel('Alcohol Content')
plt.ylabel('pH Level')
plt.show()
```
